#### Part 1: Data Cleaning and Exploration 


#### Load the dataset. Check the first 5 rows.


```python
import pandas as pd

file_path = '/Users/agrigoryan/Desktop/Masters/AUA/Data Visualization/Homeworks/Homework1/crime_data.csv'
df = pd.read_csv(file_path)
df.head(5)
```

#### Identify columns with missing values and their respective counts. Drop columns where more than 50% of the data is missing (store this version as a new dataset).

```python
missing_counts = df.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]   
print("columns with missing values and their respective counts:")
print(missing_counts)
```

```python
threshold = 0.5 * len(df)
df_cleaned = df.dropna(axis=1, thresh=threshold)
cleaned_file_path = "/Users/agrigoryan/Desktop/Masters/AUA/Data Visualization/Homeworks/Homework1/crime_data_cleaned.csv"
df_cleaned.to_csv(cleaned_file_path, index=False)
print("dropped the following columns:", set(df.columns) - set(df_cleaned.columns))
```

#### Convert the DATE OCC column to a datetime format. Extract the year, month, and day into separate columns. Create a new column for the hour using the TIME OCC column.

```python
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], errors='coerce')

df['Year'] = df['DATE OCC'].dt.year
df['Month'] = df['DATE OCC'].dt.month
df['Day'] = df['DATE OCC'].dt.day
df['TIME OCC'] = df['TIME OCC'].astype(str).str.zfill(4)  
df['Hour'] = df['TIME OCC'].str[:2].astype(int)   


df[['DATE OCC', 'Year', 'Month', 'Day', 'TIME OCC', 'Hour']].head()
```

#### Filter the dataset for crimes that occurred in 2023. Further filter crimes with the description BURGLARY in the Crm Cd Desc column.

```python
df['Year'] = df['DATE OCC'].dt.year
df_2023 = df[df['Year'] == 2023]
df_burglary_2023 = df_2023[df_2023['Crm Cd Desc'].str.upper() == "BURGLARY"]
df_burglary_2023.head()
```

#### Group the data by AREA NAME and calculate the total number of crimes and the average victim age. Sort the results by total crimes in descending order.

```python
grouped_by_area_name = df.groupby('AREA NAME').agg(Total_Crimes=('AREA NAME', 'count'), Avg_Victim_Age=('Vict Age', 'mean')).reset_index()

grouped_by_area_name_sorted = grouped_by_area_name.sort_values(by='Total_Crimes', ascending=False)
grouped_by_area_name_sorted.head()
```

#### Part 2: Further Exploration 

#### Find the top 3 most frequent Crm Cd Desc values.

```python
top_3_crimes = df['Crm Cd Desc'].value_counts().head(3).reset_index()
top_3_crimes.columns = ['Crm Cd Desc', 'Count']

top_3_crimes.head(3)
```

#### Group the data by Hour and count the number of crimes.

```python
grouped_by_hour = df.groupby('Hour').size().reset_index(name='Total_Crimes')
grouped_by_hour.head(10)
```

#### Group the data by Vict Sex and calculate: Total crimes, Average victim age.

```python
grouped_by_vict_sex = df.groupby('Vict Sex').agg(
    Total_Crimes=('Vict Sex', 'count'),
    Avg_Victim_Age=('Vict Age', 'mean')).reset_index()

grouped_by_vict_sex.head(5)
```

#### Create a new column, Severity Score, based on the following rules:
#### â€¢ Assign 5 points if a weapon was used (Weapon Used Cd is not null).
#### â€¢ Assign 3 points for crimes under BURGLARY.
#### â€¢ Assign 1 point for all other crimes.
#### â€¢ Group by AREA NAME and find the total severity score for each area.

```python
df['Severity Score'] = 1   

if 'Weapon Used Cd' in df.columns:
    df.loc[df['Weapon Used Cd'].notna(), 'Severity Score'] += 5  # add 5 points
df.loc[df['Crm Cd Desc'].str.upper() == "BURGLARY", 'Severity Score'] += 3  # add 3 points
grouped_severity = df.groupby('AREA NAME').agg(Total_Severity_Score=('Severity Score', 'sum')).reset_index()

df.head(10)
```

#### Use the LAT and LON columns to identify crimes that occurred within a specific latitude-longitude bounding box (e.g., downtown area).

```python
lat_min, lat_max = 34.02, 34.07  
lon_min, lon_max = -118.27, -118.23   

downtown_crimes = df[(df['LAT'] >= lat_min) & (df['LAT'] <= lat_max) & (df['LON'] >= lon_min) & (df['LON'] <= lon_max)]
downtown_crimes.head()
```
