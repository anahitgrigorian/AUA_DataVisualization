```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

lung_cancer_df = pd.read_csv("lung_cancer_prediction_dataset.csv")
air_pollution_df = pd.read_csv("global_air_pollution_dataset.csv")
```


# 1. Boxplot of Lung Cancer Deaths Distribution


```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=lung_cancer_df['Annual_Lung_Cancer_Deaths'])
plt.title("Boxplot of Lung Cancer Deaths Distribution")
plt.grid()
plt.show()
```

# 2. Histogram of PM2.5 AQI Values

```python
plt.figure(figsize=(8, 5))
plt.hist(air_pollution_df['PM2.5_AQI_Value'], bins=20, edgecolor="black")
plt.xlabel("PM2.5 AQI Value")
plt.ylabel("Frequency")
plt.title("Histogram of PM2.5 AQI Values")
plt.grid()
plt.show()
```

# 3. Density Plot of the Lung Cancer Mortality Rate

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(lung_cancer_df['Mortality_Rate'], fill=True)
plt.xlabel("Lung Cancer Mortality Rate")
plt.ylabel("Density")
plt.title("Density Plot of Lung Cancer Mortality Rate")
plt.grid()
plt.show()
```

# 4. Scatter Plot with Normal and Logistic Distributions

```python
normal_values = np.random.normal(loc=0, scale=1, size=100)
logistic_values = np.random.logistic(loc=0, scale=1, size=100)

plt.figure(figsize=(8, 5))
plt.scatter(normal_values, logistic_values, color='brown')
plt.xlabel("Normal Distribution")
plt.ylabel("Logistic Distribution")
plt.title("Scatter Plot of Normal vs Logistic Distribution")
plt.grid()
plt.show()
```
